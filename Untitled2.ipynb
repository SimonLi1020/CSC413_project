```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r'http\S+', '', text)  # Remove URLs
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)  # Remove non-alphanumeric characters
    return text.split()

apple_tweets_valid['tokenized_text'] = apple_tweets_valid['body'].apply(clean_text)
```

```python
max_length = max(len(tweet) for tweet in apple_tweets_valid['tokenized_text'])

def pad_tokenized_text(tokenized_text, max_len):
    padding_size = max_len - len(tokenized_text)
    if padding_size > 0:
        tokenized_text.extend(['<pad>'] * padding_size)  # Add a padding token
    return tokenized_text

# Apply padding to the tokenized text
apple_tweets_valid['padded_tokenized_text'] = apple_tweets_valid['tokenized_text'].apply(lambda x: pad_tokenized_text(x, max_length))
```

```python
def get_glove_embeddings(padded_tokenized_text, glove_model):
    embeddings = []
    for word in padded_tokenized_text:
        if word in glove_model.stoi:
            embeddings.append(glove_model[word])
        elif word == '<pad>':  # Handle the padding token
            embeddings.append(torch.zeros(50))
        else:
            embeddings.append(torch.zeros(50))  # Handling unknown words
    return embeddings

# Assuming you have the GloVe model loaded in 'glove'
apple_tweets_valid['embeddings'] = apple_tweets_valid['padded_tokenized_text'].apply(lambda x: get_glove_embeddings(x, glove))
```

```python
import torch
import torch.nn as nn
from torch.utils.data import Dataset

class TimeSeriesDataset(Dataset):
    def __init__(self, series, window_size):
        """
        series: A 2D array-like object where each row is a time step and columns are features.
        window_size: Number of time steps in each input sequence.
        """
        self.series = series
        self.window_size = window_size

    def __len__(self):
        # Adjusting length to avoid out-of-bounds indexing
        return len(self.series) - self.window_size

    def __getitem__(self, index):
        # Ensure that we don't go out of bounds
        if index + self.window_size >= len(self.series):
            raise IndexError("Index out of bounds")
        """
        Returns a tuple (input_sequence, target_value) for the given index.
        """
        # x = self.series[index:index + self.window_size, :]  # Windowed input sequence
        # y = self.series[index + self.window_size]        # Target value (next time step)
        # return torch.Tensor(x), torch.Tensor(y)
        x = self.series.iloc[index:index + self.window_size]  # Input sequence
        y = self.series.iloc[index + self.window_size]        # Target value (next time step)
        return torch.Tensor(x.values), torch.Tensor(y.values)
```

```python
import torch.nn.functional as F
import matplotlib.pyplot as plt

class TemporalPatternAttention(nn.Module):
    def __init__(self, num_units, window_size, num_filters):
        super(TemporalPatternAttention, self).__init__()
        self.num_filters = num_filters
        self.conv_filters = nn.ModuleList([nn.Conv1d(in_channels=1, out_channels=num_filters, kernel_size=window_size) for _ in range(window_size)])
        self.Wa = nn.Linear(num_filters, num_units)
        self.Wh = nn.Linear(num_units, num_units)
        self.Wv = nn.Linear(num_filters, num_units)0
    def forward(self, H, ht):
        # H: [batch_size, window_size, num_units], ht: [batch_size, num_units]
        batch_size, window_size, num_units = H.shape

        # Apply CNN filters
        HC = torch.cat([self.conv_filters[i](H[:, i:i+1, :]).permute(0, 2, 1) for i in range(window_size)], dim=1)  # [batch_size, window_size, num_filters]

        # Attention weights
        alpha = torch.sigmoid(self.Wa(HC) + self.Wh(ht).unsqueeze(1))  # [batch_size, window_size, num_units]
        vt = torch.sum(alpha * HC, dim=1)  # [batch_size, num_filters]

        # Final prediction
        ht_new = self.Wv(vt) + self.Wh(ht)  # [batch_size, num_units]
        return ht_new

```

```python
class Length(nn.Module):
    """Computes length of vectors in PyTorch."""
    def forward(self, inputs):
        return torch.sqrt(torch.sum(torch.square(inputs), dim=-1))


class Mask(nn.Module):
    """Implements the mask layer in PyTorch."""
    def forward(self, inputs):
        if isinstance(inputs, list) and len(inputs) == 2:
            inputs, mask = inputs
        else:
            x = torch.sqrt(torch.sum(torch.square(inputs), dim=-1))
            mask = F.one_hot(torch.argmax(x, dim=1), num_classes=x.size(1))

        masked = inputs * mask.unsqueeze(-1)
        return masked.view(masked.size(0), -1)


def squash(vectors, axis=-1):
    """Squashing function in PyTorch."""
    s_squared_norm = torch.sum(torch.square(vectors), axis, keepdim=True)
    scale = s_squared_norm / (1 + s_squared_norm) / torch.sqrt(s_squared_norm + 1e-7)
    return scale * vectors


class CapsuleLayer(nn.Module):
    """Capsule Layer implemented in PyTorch."""
    def __init__(self, num_capsule, dim_capsule, num_routing=3, previous_layer_size=25):
        super(CapsuleLayer, self).__init__()
        self.num_capsule = num_capsule
        self.dim_capsule = dim_capsule
        self.num_routing = num_routing
        self.W = nn.Parameter(torch.randn(num_capsule, previous_layer_size, dim_capsule))

    def forward(self, inputs):
        batch_size = inputs.size(0)
        inputs = inputs.unsqueeze(1)
        # print("Shape of inputs in CapsuleLayer before einsum:", inputs.shape)
        # print("Shape of self.W in CapsuleLayer:", self.W.shape)
        # inputs = inputs.unsqueeze(1).repeat(1, self.num_capsule, 1, 1)
        # The einsum operation
        # 'bijk, jkl -> bil' means:
        # b - batch size, i - num_capsule, j - previous_layer_size, k/l - dim_capsule
        inputs_hat = torch.einsum('bijk, jkl -> bil', [inputs, self.W])

        b = torch.zeros(batch_size, self.num_capsule, inputs.size(2)).to(inputs.device)

        for i in range(self.num_routing):
            c = F.softmax(b, dim=1)
            outputs = squash(torch.einsum('bij,bjk->bik', [c, inputs_hat]))
            if i < self.num_routing - 1:
                b = torch.einsum('bij,bjk->bik', [outputs, inputs_hat])

        return outputs


# Primary Capsule Layer class
# class PrimaryCap(nn.Module):
#     def __init__(self, dim_capsule, n_channels, kernel_size, stride, padding, hidden_size):
#         super(PrimaryCap, self).__init__()
#         self.dim_capsule = dim_capsule
#         self.conv = nn.Conv1d(in_channels=hidden_size, out_channels=dim_capsule * n_channels,
#                               kernel_size=kernel_size, stride=stride, padding=padding)
        
#     def forward(self, x):
#         # Reshape the input for Conv1d
#         # Assuming x is of shape [batch_size, 1, hidden_size, 1]
#         # x = x.squeeze(-1)  # Now x is of shape [batch_size, hidden_size, 1]
#         output = self.conv(x)
#         output = output.view(x.size(0), -1, self.dim_capsule)
#         return squash(output)
    
class PrimaryCap(nn.Module):
    def __init__(self, dim_capsule, n_channels, kernel_size, stride, padding, hidden_size):
        super(PrimaryCap, self).__init__()
        self.dim_capsule = dim_capsule
        # Modify the out_channels to be dim_capsule * n_channels
        # Ensure that dim_capsule * n_channels equals the expected size (8 in this case)
        self.conv = nn.Conv1d(in_channels=hidden_size, out_channels=dim_capsule * n_channels,
                              kernel_size=kernel_size, stride=stride, padding=padding)

    def forward(self, x):
        output = self.conv(x)
        # Modify the reshape operation to ensure the last dimension is dim_capsule
        output = output.view(x.size(0), -1, self.dim_capsule)
        return squash(output)

```

```python
class RNNWithTemporalPatternAttention(nn.Module):
    def __init__(self, input_size, hidden_size, window_size, num_filters, num_capsule, dim_capsule, output_size):
        super(RNNWithTemporalPatternAttention, self).__init__()
        self.rnn = nn.LSTM(input_size=input_size, hidden_size=hidden_size, batch_first=True)
        self.attention = TemporalPatternAttention(hidden_size, window_size, num_filters)
        # self.fc = nn.Linear(hidden_size, output_size)
        # add capsule layer
        self.primary_caps = PrimaryCap(dim_capsule=hidden_size, n_channels=1, 
                                       kernel_size=1, stride=1, padding=0, hidden_size=hidden_size)
        self.capsule_layer = CapsuleLayer(num_capsule=num_capsule, dim_capsule=dim_capsule)
        self.fc = nn.Linear(num_capsule * dim_capsule, output_size)

    def forward(self, x):
        # x: [batch_size, seq_len, input_size]
        batch_size, seq_len, _ = x.shape

        # RNN
        rnn_out, (ht, _) = self.rnn(x)  # [batch_size, seq_len, hidden_size], ht: [1, batch_size, hidden_size]
        ht = ht.squeeze(0)  # [batch_size, hidden_size]
        # print("Shape after RNN:", rnn_out.shape)

        # Temporal Pattern Attention
        attn_out = self.attention(rnn_out[:, -window_size:], ht)  # [batch_size, hidden_size]
        # print("Shape after Temporal Pattern Attention:", attn_out.shape)


        # Reshape for the Primary Capsule Layer
        # Reshape attn_out to [batch_size, hidden_size, 1] for Conv1d
        caps_input = attn_out.unsqueeze(-1)
        # print("Shape before Primary Capsule Layer:", caps_input.shape)

        # Primary Capsule Layer
        caps_out = self.primary_caps(caps_input)
        # print("Shape after Primary Capsule Layer:", caps_out.shape)

        # Capsule Layer
        caps_out = self.capsule_layer(caps_out)
    
        # Flatten the output for the fully connected layer
        # caps_out = caps_out.view(batch_size, -1)
        caps_out = caps_out.view(caps_out.size(0), -1)
        # print("Shape before the final fully connected layer:", caps_out.shape)

        # Final Output
        out = self.fc(caps_out)  # [batch_size, output_size]
        # print("Final output shape:", out.shape)
        
        return out

# Model initialization
num_filters = 25  # Example number of filters
num_capsule = 5
dim_capsule = 8
n_channels = 1
# used to be input_size=50
model = RNNWithTemporalPatternAttention(input_size=2, hidden_size=25, window_size=window_size, 
                                        num_filters=num_filters, num_capsule=num_capsule, 
                                        dim_capsule=dim_capsule, output_size=1)
```

```python
import torch.optim as optim

def train_model(model, train_data, val_data, learning_rate=0.001, batch_size=100, num_epochs=10, plot_every=50, plot=True):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)

    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)

    train_loader = DataLoader(train_data, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(val_data, batch_size=batch_size, shuffle=False)

    iters, train_loss, val_loss = [], [], []
    iter_count = 0
    best_val_loss = float("inf")

    try:
        for epoch in range(num_epochs):
            model.train()
            for inputs, targets in train_loader:
                inputs, targets = inputs.to(device), targets.to(device)

                outputs = model(inputs)
                loss = criterion(outputs, targets)

                optimizer.zero_grad()
                loss.backward()
                optimizer.step()

                iter_count += 1
                if iter_count % plot_every == 0:
                    train_loss.append(loss.item())
                    print(f"Iteration {iter_count}, Training Loss: {loss.item()}")

            model.eval()
            with torch.no_grad():
                total_val_loss = 0
                for val_inputs, val_targets in val_loader:
                    val_inputs, val_targets = val_inputs.to(device), val_targets.to(device)
                    val_outputs = model(val_inputs)
                    val_loss_value = criterion(val_outputs, val_targets)
                    total_val_loss += val_loss_value.item()

                avg_val_loss = total_val_loss / len(val_loader)
                val_loss.append(avg_val_loss)
                print(f"Epoch {epoch}, Validation Loss: {avg_val_loss}")

                if avg_val_loss < best_val_loss:
                    best_val_loss = avg_val_loss
                    torch.save(model.state_dict(), "best_model.pth")

    finally:
        if plot:
            plt.figure()
            plt.plot(train_loss, label='Training Loss')
            plt.plot(val_loss, label='Validation Loss')
            plt.xlabel('Iterations')
            plt.ylabel('Loss')
            plt.title('Loss over Iterations')
            plt.legend()
            plt.show()


train_model(model, train_dataset, val_dataset, learning_rate=0.002, batch_size=3, num_epochs=5, plot_every=100, plot=True)
```
